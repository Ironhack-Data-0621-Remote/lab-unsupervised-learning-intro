```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')   
df.head()
```

```python
df.info()
```

```python
df = df.set_index('id')
df.head()
```

```python
X = df.select_dtypes(np.number)
```

```python
# K-Means is a distance based algorithm: we need to scale / normalize:

X_scaled = StandardScaler().fit_transform(X)

pd.DataFrame(X_scaled).head()
```

```python
#copying the dataframe

X1 = X_scaled.copy()
X2 = X_scaled.copy()
X3 = X_scaled.copy()
```

```python
# I tried to calculate the silhouette score to choose the optimal k number 
# but it took too much time and did'nt give me any result.

K = range(3,6)

silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    silhouette.append(silhouette_score(X_scaled, kmeans.predict(X_scaled)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.show()
```

So I will run the model with random K numbers.

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
```

```python
# Check the size of the clusters
clusters = kmeans.predict(X_scaled)
pd.Series(clusters).value_counts().sort_index()
```

```python
# "performance metric"
kmeans.inertia_
```

```python
def k_means_model(df,n_clusters):
    
    kmeans = KMeans(n_clusters, random_state=42)
    kmeans.fit(df)
    
    # Check the size of the clusters
    clusters = kmeans.predict(df)
    print(pd.Series(clusters).value_counts().sort_index())
    
    # "performance metric"
    print(kmeans.inertia_)
```
