# Billboard

## Importing libraries

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from bs4 import BeautifulSoup
import requests
from sklearn.metrics import silhouette_score
```

## Importing dataset

```python
data = "https://www.billboard.com/charts/billboard-global-200"
```

```python
hot_songs = requests.get(data)
print("hot-100 songs: status",hot_songs.status_code)
```

```python
soup = BeautifulSoup(hot_songs.content, 'html.parser')
```

```python
# song titles
song_titles = soup.find_all("span", class_="chart-element__information__song text--truncate color--primary")

# artists
artists = soup.find_all("span", class_="chart-element__information__artist text--truncate color--secondary")

# last week
last_week = soup.find_all("span", class_= "chart-element__meta text--center color--secondary text--last")

# peak rank
peak_rank = soup.find_all("span", class_= "chart-element__meta text--center color--secondary text--peak")

# weeks on chart
weeks_on_chart = soup.find_all("span", class_= "chart-element__meta text--center color--secondary text--week")
```

```python
for i in [song_titles, artists, last_week, peak_rank, weeks_on_chart]:
    for j in range(len(i)):
        i[j] = i[j].getText()
```

```python
billboard = pd.DataFrame(
    {"song_titles": song_titles,
     "artists": artists,
     "last_week": last_week,
     "peak_rank": peak_rank,
     "weeks_on_chart": weeks_on_chart})
billboard
```

```python
def song_in_top200(): 
    song = input("What song do you want to check? ")
    if song in song_titles: 
        print('Yes, song is in top 200')
    else:    
        print('Not a hit. Try an other song')
```

```python
song_in_top200()
```

# Spotify

```python
data = pd.read_csv('playlist_spotify.csv')
data
```

```python
playlist = data.select_dtypes(include = np.number)
playlist
```

```python
playlist = playlist.drop(['Duration (ms)', 'Popularity'], axis = 1)
playlist
```

```python
X_prep = StandardScaler().fit_transform(playlist)

pd.DataFrame(X_prep).head()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_prep)
```

```python
clusters = kmeans.predict(X_prep)
clusters
```

```python
pd.Series(clusters).value_counts().sort_index()
```

```python
kmeans.inertia_
```

# Choosing the best K

## Inertia

```python
kmeans = KMeans(n_clusters=3, random_state=1234, verbose=1, n_init=1)
kmeans.fit(X_prep)
kmeans.inertia_
```

```python
K = range(1, 20,2)

inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=1234)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.show()
```

## Silhouette Score

```python
K = range(2, 20)

silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.show()
```
